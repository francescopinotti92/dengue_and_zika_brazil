```python
#== prepare observations

#= read different data sources
locs = pd.read_csv( '../data/uf_locations.csv' ).set_index( 'code' ) # contains spatial locations of each state (uf)
dsh  = pd.read_csv( '../data/mca_h_fit.csv' ) # contains mean hosp age by year and state
dsc  = pd.read_csv( '../data/mca_c_fit.csv' ) # contains mean case age by year and state

#= format data
data = { 'code': [], 'age_h': [], 'age_c': [], 'x': [], 'y': [], 't': [] }

for ic, code in enumerate( codes ):
    for iy, year in enumerate( range( 2001, 2020 ) ):
        
        data['code'].append( code ) # uf code
        data['t'].append( year ) # year
        data['x'].append( locs.loc[code,'x'] ) # x uf coord
        data['y'].append( locs.loc[code,'y'] ) # y uf coord
        data['age_h'].append( dsh.loc[iy,code] ) # mean hosp age
        data['age_c'].append( dsc.loc[iy,code] ) # mean case age
    
data = pd.DataFrame( data )
```

```python
#== prepare training data

#= SELECT either case of hospitalisation data

#col_fit = 'h' # analyse hospitalisation data
col_fit = 'c' # analyse hospitalisation data

data_train = data.loc[ data['t'] <= 2015 ].copy()
data_train['age_h_log'] = np.log( data_train['age_h'] )
data_train['age_c_log'] = np.log( data_train['age_c'] )

X_train = data_train[ [ 'x', 'y', 't' ] ].values

if col_fit == 'h':
    y_train = data_train[ 'age_h_log' ].values
elif col_fit == 'c':
    y_train = data_train[ 'age_c_log' ].values
else:
    raise ValueError
    
y_train_mean = y_train.mean()
y_train_1 = y_train - y_train_mean

#== setup kernel and creat gp object

#= create kernel object
spt_kernel =  2. * RBF( length_scale = [5.,5.,1.] )
noise_kernel = WhiteKernel( noise_level=0.1) 
kernel = spt_kernel + noise_kernel 

#= create gp object
gp = GaussianProcessRegressor( kernel = kernel, n_restarts_optimizer=10, normalize_y=False )

#== fit gp

gp.fit( X_train, y_train_1 )
print( gp.kernel_ )
print( gp.log_marginal_likelihood_value_ )
```

### Analyse residuals

```python
#=== calculate age shift Δ for each uf

n_samples = 5000

if col_fit == 'c':
    colnamedata = 'age_c'
elif col_fit == 'h':
    colnamedata = 'age_h'

Δss = {}

for code in codes:
    
    X_pred = np.vstack( [ [locs.loc[code,'x']]*5, 
                          [locs.loc[code,'y']]*5,
                          [2013, 2014, 2015, 2018, 2019]] ).T
    
    y_post = np.exp( gp.sample_y( X = X_pred, n_samples = n_samples ) + y_train_mean )
    age_pre  = y_post[:3].mean( axis = 0 )
    age_post = y_post[3:].mean( axis = 0 )
    
    age_pre_data  = data.loc[ ( data['code'] == code ) & ( data['t'].isin( [ 2013, 2014, 2015 ] ) ), colnamedata ].mean() 

    Δ1 = age_post / age_pre_data - 1 # use age_pre from data
    Δss[code] = Δ1

Δss = { col_fit: pd.DataFrame( Δss ) }

#== calculate residuals and correlation coefficient


#= calculate age shift Δ from data
Δss_data = {}
tmp = []

for code in codes:
    age_pre  = data.loc[ ( data['code'] == code ) & ( data['t'].isin( [ 2013, 2014, 2015 ] ) ), colnamedata ].mean()    
    age_post = data.loc[ ( data['code'] == code ) & ( data['t'].isin( [ 2018, 2019 ] ) ), colnamedata ].mean()
    Δ = age_post / age_pre - 1
    tmp.append( Δ )
        
    Δss_data[col_fit] = np.array( tmp )

#= residuals: Δ data - Δ GP
resids = {
    col_fit: Δss_data[col_fit] - Δss[col_fit],
}

# correlation coefficient
corrs = {
    col_fit: [ spearmanr( Δss[col_fit].loc[i], zikv_ar )[0] for i in range( len( Δss[col_fit] ) ) ],
}
```
